## Exploratory Data Analysis (EDA)

##### Exploratory Data Analysis was performed to identify patterns and relationships between "Quality","sweetness" features. Various visualization techniques were used to better understand the dataset.

```python
plt.scatter(df['Brix (Sweetness)'], df['Quality (1-5)'])
plt.xlabel("Sweetness (Brix)")
plt.ylabel("Quality")
plt.title("Sweetness vs Orange Quality")
plt.show()
```

```python
# It shows how many oranges fall into each quality level ranging from
sns.countplot(x='Quality (1-5)', data=df)
plt.title("Quality Distribution")
plt.show()
```

##### Scatter plots were used to analyze the relationship between different features of oranges and their quality rating. Each scatter plot compares one feature (such as size, weight, sweetness, acidity, softness, harvest time, and ripeness) with the orange quality level. This helps in understanding whether changes in these features influence the quality of oranges.


 Size (cm)
Size increase ayina quality score sometimes same ga untundi. Kabatti size alone orange quality ni strong ga decide cheyyadu.

Weight (g)
Weight change ayina quality lo clear pattern kanipinchakapovachu. Idi weight quality ki strong factor kaakapovachu ani suggest chestundi.

 Brix (Sweetness)
Sweetness ekkuva unna oranges lo quality score kuda ekkuva undachu. Idi sweetness quality ni influence cheyyachu ani indicate chestundi.

 pH (Acidity)
Acidity level change ayina quality lo konni variations kanipinchachu. Moderate acidity better quality ki sambandham undachu.

 Softness (1-5)
Softness level change ayite quality score kuda change ayye chance undi. Proper softness unna oranges better quality indicate cheyyachu.

 HarvestTime (days)
Harvest time different ga unna quality lo konni differences kanipinchachu. Correct harvest time better orange quality ki help cheyyachu.

 Ripeness (1-5)
Ripeness increase ayite quality score kuda increase ayye tendency undachu. Fully ripened oranges generally higher quality ni indicate chestayi.

```python
features = [
    "Size (cm)",
    "Weight (g)",
    "Brix (Sweetness)",
    "pH (Acidity)",
    "Softness (1-5)",
    "HarvestTime (days)",
    "Ripeness (1-5)"
]

for feature in features:
    sns.scatterplot(x=df[feature], y=df["Quality (1-5)"])
    plt.title(f"{feature} vs Orange Quality")
    plt.xlabel(feature)
    plt.ylabel("Quality (1-5)")
    plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.histplot(df["Quality (1-5)"], kde=True)
plt.show()
```

##### In this heatmap, different features such as size, weight, sweetness (Brix), acidity (pH), softness, harvest time, ripeness, and quality are compared with each other. The color intensity represents the strength of the relationship between variables.

```python
# correlation heatmap
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation")
plt.show()
```

3. The correlation heatmap highlights the relationships between numerical features, helping identify which factors may influence the quality of oranges more significantly.
